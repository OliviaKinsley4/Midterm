```python
import pandas as pd
df = pd.read_csv('amazon-orders.csv')
df.head()
```

```python
print("Table dimensions are", df.shape)
```

```python
df = df.fillna(0)
df.head()
```

```python
df["Total Charged"] = df["Total Charged"].str.replace('$','').astype(float)
df.head()
```

```python
df["Tax Charged"] = df["Tax Charged"].str.replace('$','').astype(float)
df.head()
```

```python
print("Sum of Amazon purchases is", df["Total Charged"].sum())
print("Mean of Amazon purchases is", df["Total Charged"].mean())
print("Median of Amazon purchases is", df["Total Charged"].median())
print("Most expensive Amazon purchases was", df["Total Charged"].max())
print("Least expensive Amazon purchases was", df["Total Charged"].min())
print("Standard deviation of total charged is", df["Total Charged"].std())
```

```python
print("Sum of tax charged is", df["Tax Charged"].sum())
print("Mean of tax charged is",df["Tax Charged"].mean())
print("Median of tax charged is", df["Tax Charged"].median())
print("Maximum of tax charged is", df["Tax Charged"].max( ))
print("Minimum of tax charged is", df["Tax Charged"].min())
print("Standard deviation of tax charged is", df["Tax Charged"].std())
```

```python
df['Order Month'] = df['Order Date'].dt.to_period("M")
df
```

```python
Monthly_Spending = df.groupby("Order Month").sum("Total Charged")
Monthly_Spending
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array(Monthly_Spending['Total Charged'])
mylabels = Monthly_Spending.index

plt.pie(y, labels = mylabels)
plt.title("Monthly Spending")
plt.legend(title = "Month", bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array(Monthly_Spending['Tax Charged'])
mylabels = Monthly_Spending.index
plt.pie(y, labels = mylabels)
plt.title("Monthly Taxes")
plt.legend(title = "Month", bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show
```

```python
Monthly_Spending['Total Charged'].plot.bar(figsize=(20, 10), color='pink')
plt.title("2019 Orders by Month")
```

```python
daily_orders = df.groupby('Order Date').sum()["Total Charged"]
daily_orders.head()
```

```python
daily_orders.plot.bar(figsize=(20, 10), color='pink')
plt.title("2019 Orders")
```
